```python
import collections
import glob
import os
import re
import shutil

import numpy as np
import pandas as pd

from PIL import Image
```

1.データの確認

```python
# メインにするデータセットの中身の確認
dir_path_main = "../data/raw/Weather Image Recognition"

# メインのデータセットのフォルダ名を取得
list_dir_main = os.listdir(dir_path_main)
```

```python
# フォルダ名の確認
list_dir_main
```

```python
# クラスの分布の確認
# クラスごとにフォルダ分けされているので、フォルダの中身のファイル数を取得

list_class_path_main = []
list_class_freq_main = []

for dir in list_dir_main:
    target_path = os.path.join(dir_path_main, dir)

    print(target_path)

    list_class_path_main.append(target_path)
    list_class_freq_main.append((dir, len(os.listdir(target_path))))

```

```python
# メインのデータセットのクラス分布の確認
list_class_freq_main
```

```python
# 画像サイズの確認

def check_img_size(dir_path):

    l =[]

    list_filename = os.listdir(dir_path)

    for file_name in list_filename:

        file_path = os.path.join(dir_path, file_name)

        img = Image.open(file_path)

        width, height = img.size

        l.append((width, height))

    # return file_path

    return l
```

```python
# メインの画像サイズを確認

list_img_size_main = []

for path in list_class_path_main:
    # print(path)

    list_img_size_main.append(check_img_size(dir_path=path))
```

```python
print(list_img_size_main)

# メインの画像サイズが揃っていないことを確認
```

```python
# サブのデータセットの中身の確認
dir_path_sub = "../data/raw/Weather Dataset for Image Classification"

list_files_sub = os.listdir(dir_path_sub)
```

```python
# ファイルの数を確認、抜け漏れとかがないか
len(list_files_sub)
```

```python
list_filename = [i.split(".")[0] for i in list_files_sub]
```

```python
# スプリットがうまくうごいているか確認
list_filename
```

```python
# 画像ファイルの数の確認
# len(list_filename)
len(list_files_sub)
```

```python
# 元データのカテゴリーを取り出す。
# 元データのファイルの命名規則が、カテゴリ名＋ナンバリング、なので、カテゴリ名だけをとりだす

list_cat = []

for i in list_files_sub:
# for i, item in enumerate(list_filename):

    matched = re.match(r"([a-zA-Z]+)(\d+)", i)
    # matched = re.match(r"([a-zA-Z]+)(\d+)", item)


    # # print(i)

    # if matched is None:
    #     print(f"None is:{i}." , "Matched:", matched)
    # # print(matched.group(1))

    list_cat.append(matched.group(1))
```

```python
# サブのデータセットのファイルの数が減ってないか確認
len(list_files_sub) == len(list_cat)
```

```python
list_img_size_sub = check_img_size(dir_path_sub)
```

```python
print(list_img_size_sub)

# サブの画像サイズが揃っていないことを確認
```

```python
# サブのカテゴリの分布を確認
list_class_freq_sub = collections.Counter(list_cat)
print(list_class_freq_sub)
```
