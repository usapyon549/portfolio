```python
import glob
import os

import pandas as pd
```

1.ファイル読み込み

1.1分類前のyahooコメントの読み込み

```python
df_yahoo_raw = pd.read_csv("../data/yahoo_comments_renumbering.csv")
```

```python
df_yahoo_raw
```

1.ラベル付けされたyahooコメントの読み込み

```python

# バッチで処理した複数ファイルのラベリング結果の保存先ディレクトリ

# ファイル一覧取得
list_file_names = glob.glob("../data/gemini_response/*.csv")
```

```python
# ファイルの読み込み

df_yahoo_labeled = pd.DataFrame()

for fp in list_file_names:

    df = pd.read_csv(fp)

    df_yahoo_labeled = pd.concat([df_yahoo_labeled,df])

# df.head(5)
```

```python
# 見やすくするためidでsort
df_yahoo_labeled = df_yahoo_labeled.sort_values("id")
```

1.3分類前のyahooコメントにラベルをマージ

```python
# ラベル付け後のyahooコメントには日付がないため、マージする。
# プロンプトの安定性重視のため、最低限の出力にするためにラベリングのさいには日付を含めなかった

df_yahoo_merged = pd.merge(df_yahoo_raw, df_yahoo_labeled, on="id", how="inner")
```

```python
# マージができてるか確認
df_yahoo_merged.head(5)
```

2.分析

```python
# 分析に使用するカラムを抜き出す
use_cols = ["article_id","date","label"]

# use_cols = ["id","article_id","date","label"]

df_analysis = df_yahoo_merged[use_cols].copy()
```

```python
df_analysis
```

2.1ラベルの分布の確認

```python
# 件数分布
df_analysis.groupby("article_id")["label"].value_counts()
```

```python
# 記事Cの強い批判が思ったより多い？
```

```python
# 比較のために正規化
df_analysis.groupby("article_id")["label"].value_counts(normalize=True)
```

2.2コメント性質の変化

```python
# 記事ごと、日毎、の推移を計算
df_analysis.groupby(["article_id","date"])["label"].value_counts(normalize=True)
```

```python
df_proportion = df_analysis.groupby(["article_id","date"])["label"].value_counts(normalize=True).reset_index()
```
